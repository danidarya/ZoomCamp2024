```python
question_3 = '''
select count(*) from green_tripdata_trip
where date(lpep_pickup_datetime) = '2019-09-18' and date(lpep_dropoff_datetime) = '2019-09-18'
'''

pd.read_sql(question_3, con=engine)
```

```python
question_3 = '''
select  max(trip_distance) as trip_distance, 
date(lpep_pickup_datetime) as trip_date
from green_tripdata_trip
group by date(lpep_pickup_datetime)
order by trip_distance desc
limit 3
'''

pd.read_sql(question_3, con=engine)
```

```python
question_5 = '''
select sum(total_amount) as total_sum,
       "Borough"
from green_tripdata_trip t
join taxi_zone z on t."PULocationID" = z."LocationID"
where date(lpep_pickup_datetime) = '2019-09-18' and "PULocationID" is not Null
group by "Borough"
having sum(total_amount) > 50000
order by total_sum desc
'''

pd.read_sql(question_5, con=engine)
```

```python
question_6 = '''
select
   z_dropoff."Zone" as dropoff_zone,
   MAX("tip_amount") as largest_tip
from green_tripdata_trip t
join taxi_zone z_pickup on t."PULocationID" = z_pickup."LocationID"
join taxi_zone z_dropoff on t."DOLocationID" = z_dropoff."LocationID"
where DATE(t."lpep_pickup_datetime") >= '2019-09-01'
  and DATE(t."lpep_pickup_datetime") <= '2019-09-30'
  and z_pickup."Zone" = 'Astoria'
group by z_dropoff."Zone"
order by largest_tip desc
limit 1;
'''

pd.read_sql(question_6, con=engine)
```
